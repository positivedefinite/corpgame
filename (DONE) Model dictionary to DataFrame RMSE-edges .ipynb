### All together

```python
alldict = defaultdict()
for k, p in enumerate(pops):
    instance = defaultdict()
    instance['nonzero']=p['error']!=11.056420959987207
    instance['error']=p['error']
    instance['alpha']=p['alpha']
    state = defaultdict()
    for i, arr in enumerate(p['starting_state']):
        for j, arg in enumerate(arr):
            state[f"state_{i}_{j}"]=arg
    strategies = defaultdict()
    for i, arr in enumerate(p['strategies']):
        for j, arg in enumerate(arr[0]):
            strategies[f"strategy_{i}_{j}"]=arg
    f = itertools.combinations(list(range(27)),2)
    topology = defaultdict()
    for e in f:
        topology[f"{e}"]=0
    for i, tup in enumerate(p['topology']):
        topology[f"{tup}"]=1
    instance.update(state)
    instance.update(strategies)
    instance.update(topology)
    alldict[k] = instance.values()
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(alldict, orient='index', columns=columns)
#df = df[df['nonzero']==True]
df
```

```python
df['gain']=11.056420959987207-df['error']
gains = df['gain'].values
max_gain = max(gains)
max_gain
```

```python
def fam(x):
    v = 16-x*16/max_gain
    return hex(int(v))[2:3]
fam(0.32)
```

```python
dff = df.drop(columns=['nonzero','error'])
#y = df['nonzero'].values

cmap = ['#000000' if i==0 else f'#{fam(i)}{fam(i)}0000' for i in gains]
x = dff.values
cmap
```

```python
#run PCA to find most important distances
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=2)
pca.fit(x)
X = pca.transform(x)
# plot
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1],c=cmap)
```
