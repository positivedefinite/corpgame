```python
from etl import population
df = population.to_df()
df
```

```python
states = 0 
strategies = 0
edges = 0
for e in list(df.columns):
    if '(' in e:
        edges+=1
    if 'stra' in e:
        strategies +=1
    if 'state' in e:
        states += 1

print(states, strategies, edges)
```

```python
dfc = df.corr()
dfc['error'].sort_values()
```

```python
dfc['error'].sort_values()[350:400]
```

```python
df['gain']=1.65-df['error']
gains = df['gain'].values
max_gain = max(gains)
max_gain

dff = df.drop(columns=['nonzero','error'])
#y = df['nonzero'].values

def fam(x):
    v = 16-x*16/max_gain
    return hex(int(v))[2:3]
fam(0.32)

cmap = ['#000000' if i==0 else f'#{fam(i)}{fam(i)}0000' for i in gains]
x = dff.values
cmap
```

```python
#run PCA to find most important distances
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=2)
pca.fit(x)
X = pca.transform(x)
# plot
fig, ax = plt.subplots()
ax.scatter(X[:,0], X[:,1],c=cmap)
```

```python
for i in range(len(X[:,0])):
    print(X[i,:], cmap[i])
```
